```python
import pandas as pd

orders = pd.read_csv('/content/innomatics/orders.csv')
print("Orders loaded. Shape:", orders.shape)

```

```python

users = pd.read_json('/content/innomatics/users.json')
print("Users loaded. Shape:", users.shape)
```

```python
import sqlite3

conn = sqlite3.connect(':memory:')

with open('/content/innomatics/restaurants.sql', 'r') as f:
    sql_script = f.read()
conn.executescript(sql_script)


restaurants = pd.read_sql_query("SELECT * FROM restaurants", conn)
print("Restaurants loaded. Shape:", restaurants.shape)
```

```python

merged_data = orders.merge(users, on='user_id', how='left')


final_df = merged_data.merge(restaurants, on='restaurant_id', how='left', suffixes=('_order', '_rest'))
```

```python

final_df.to_csv('/content/innomatics/final_food_delivery_dataset.csv', index=False)
print("Final dataset created successfully!")
```

```python
import pandas as pd
my_data = pd.read_csv('/content/innomatics/final_food_delivery_dataset.csv')


only_gold_people = my_data[my_data['membership'] == 'Gold']


city_gold_vault = only_gold_people.groupby('city')['total_amount'].sum()
print("Total Gold Money by the City: ")
print(city_gold_vault.sort_values(ascending=False))
```

```python

food_price_report = my_data.groupby('cuisine')['total_amount'].mean()
print("Average Price per the  Cuisine:")
print(food_price_report.sort_values(ascending=False))
```

```python

user_piggy_banks = my_data.groupby('user_id')['total_amount'].sum()
big_spenders = user_piggy_banks[user_piggy_banks > 1000]
print(f"Number of big spenders: {len(big_spenders)}")
```

```python
rating_boxes = [3.0, 3.5, 4.0, 4.5, 5.0]
box_labels = ['3.0-3.5', '3.6-4.0', '4.1-4.5', '4.6-5.0']

my_data['star_rating_group'] = pd.cut(my_data['rating'], bins=rating_boxes, labels=box_labels)
star_power_revenue = my_data.groupby('star_rating_group')['total_amount'].sum()
print(star_power_revenue.sort_values(ascending=False))
```

```python

gold_birds = my_data[my_data['membership'] == 'Gold']
gold_city_averages = gold_birds.groupby('city')['total_amount'].mean()
print("Average Gold Bill by City:")
print(gold_city_averages.sort_values(ascending=False))
```

```python

cuisine_variety = my_data.groupby('cuisine')['restaurant_id'].nunique()
print("Number of Unique Restaurants per Cuisine:")
print(cuisine_variety.sort_values())

cuisine_money = my_data.groupby('cuisine')['total_amount'].sum()
print("\nTotal Money per Cuisine:")
print(cuisine_money)
```

```python

all_orders_count = len(my_data)
gold_orders_count = len(my_data[my_data['membership'] == 'Gold'])

gold_percentage = (gold_orders_count / all_orders_count) * 100
print(f"Percentage of the  Gold orders: {(gold_percentage)}%")
```

```python

shop_stats = my_data.groupby(['restaurant_id', 'restaurant_name_order']).agg(
    total_visits=('order_id', 'count'),
    avg_bill=('total_amount', 'mean')
)
small_shops = shop_stats[shop_stats['total_visits'] < 20]
print("Top expensive small restaurants:")
print(small_shops.sort_values(by='avg_bill', ascending=False).head(5))
```

```python

mega_combo = my_data.groupby(['membership', 'cuisine'])['total_amount'].sum()
print("Money made by different combinations:")
print(mega_combo.sort_values(ascending=False))
```

```python

my_data['order_date'] = pd.to_datetime(my_data['order_date'])
my_data['year_quarter'] = my_data['order_date'].dt.to_period('Q')
quarterly_money = my_data[my_data['order_date'].dt.year == 2023].groupby('year_quarter')['total_amount'].sum()
print("Revenue by Quarter:")
print(quarterly_money.sort_values(ascending=False))
```

```python
delivery_data = pd.read_csv('/content/innomatics/final_food_delivery_dataset.csv')
gold_club_orders = delivery_data[delivery_data['membership'] == 'Gold']
total_gold = len(gold_club_orders)
print('the total orders by the gold users :',total_gold)


hyderabad_orders = delivery_data[delivery_data['city'] == 'Hyderabad']
hyderabad_cash_flow = hyderabad_orders['total_amount'].sum()
print("Hyderabad Total Revenue: ",(hyderabad_cash_flow))


unique_hungry_users = delivery_data['user_id'].nunique()
print(f"Number of distinct users: {unique_hungry_users}")


gold_only = delivery_data[delivery_data['membership'] == 'Gold']
avg_gold_bill = gold_only['total_amount'].mean()
print(f"Average Gold order value: {(avg_gold_bill)}")


top_rated_orders = delivery_data[delivery_data['rating'] >= 4.5]
high_rating_count = len(top_rated_orders)
print(f"Orders with rating 4.5 or more: {high_rating_count}")



gold_members = delivery_data[delivery_data['membership'] == 'Gold']
city_money_rank = gold_members.groupby('city')['total_amount'].sum()
best_gold_city = city_money_rank.idxmax()
best_city_gold_orders = gold_members[gold_members['city'] == best_gold_city]
print(f"Top Gold City: {best_gold_city}")
print(f"Orders in that city: {len(best_city_gold_orders)}")
```
